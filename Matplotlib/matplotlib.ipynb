```python
df = pd.read_csv('Used_Bikes.csv')
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['brand'].value_counts()                
```

```python
brand_name =list(df['brand'].value_counts().keys())
brand_name
```

```python
no_of_count =list(df['brand'].value_counts().values)
no_of_count
```

```python
plt.bar(brand_name,no_of_count)
plt.show()
```

```python
brand_name =list(df['brand'].value_counts().keys())
no_of_count =list(df['brand'].value_counts().values)
plt.bar(brand_name,no_of_count)
plt.xticks(rotation=75)
#plt.yticks(rotation=75)
plt.show()
```

```python
brand_name =list(df['brand'].value_counts().keys())
no_of_count =list(df['brand'].value_counts().values)
plt.bar(brand_name,no_of_count,color='green')
plt.xticks(rotation=75)
#plt.yticks(rotation=75)
plt.show()
```

```python
brand_name =list(df['brand'].value_counts().keys())
no_of_count =list(df['brand'].value_counts().values)
plt.bar(brand_name,no_of_count,color='yellow')
plt.xticks(rotation=75)
#plt.yticks(rotation=75)
plt.show()
```

```python
plt.figure(figsize=(10,4))   # (width, height )
brand_name =list(df['brand'].value_counts().keys())
no_of_count =list(df['brand'].value_counts().values)
plt.bar(brand_name,no_of_count,color='green',ec="black")
plt.xticks(rotation=75)
#plt.yticks(rotation=75)
plt.savefig('brandpopulaton')   # brandpopulation
plt.show()
```
