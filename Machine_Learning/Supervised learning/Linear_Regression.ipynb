# Splitting data dependent and independent

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)  # test size 20% 
```

```python
# importing linerarregression algorithm class
from sklearn.linear_model import LinearRegression
```

```python
# instance creation
lnr = LinearRegression()
```

```python
## Algorithm train
lnr.fit(x_train,y_train)
```

```python
## Model evaluation
print("Training score  ",lnr.score(x_train,y_train) * 100)
print("Testing score  ",lnr.score(x_test,y_test) * 100)
```

```python
x_test.shape  
# total 1462
```

```python
### Prediction or testing phase 


# Overfit condition ==>  Trainig - High   and Testing - Low
# Underfit condition ==>  Trainig - Low   and Testing - High
# Generalized condition ==>  if Training is greater then Testing with 5 number


prediction = lnr.predict(x_test)
```
