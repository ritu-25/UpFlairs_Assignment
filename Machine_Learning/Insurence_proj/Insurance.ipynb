# Algorithm

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
from sklearn.svm import SVC,SVR
from sklearn.neighbors import KNeighborsClassifier,KNeighborsRegressor
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
lnr = LinearRegression()
dtr = DecisionTreeRegressor()
rdr = RandomForestRegressor(n_estimators=101,min_samples_split=20,min_samples_leaf=10)


### Training algorithm

lnr.fit(x_train,y_train)
print("LinearRegression is successfully train ! ..")


dtr.fit(x_train,y_train)
print("DecisionTreeRegression is successfully train ! ..")

rdr.fit(x_train,y_train)
print("RendomForestRegression is successfully train ! ..")

```

```python
print("LinearRegressor Training Score : ",lnr.score(x_train,y_train)*100)
print("LinearRegressor Training Score : ",lnr.score(x_train,y_train)*100)
print()


print("DecisionTreeRegression Training Score : ",dtr.score(x_train,y_train)*100)
print("DecisionTreeRegression Training Score : ",dtr.score(x_train,y_train)*100)
print()


print("RendomForestRegression Training Score : ",rdr.score(x_train,y_train)*100)
print("RendomForestRegression Training Score : ",rdr.score(x_train,y_train)*100)
print()
```
