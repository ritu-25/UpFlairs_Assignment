```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot  as plt
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df = sns.load_dataset('tips')
df
```

# relational plot

```python
# replot() ==>  relational plot  ==>  for numericals
# catplot ==> categorical plot ==> for  numerical or categorical
```

```python
sns.relplot(x='tips',y='total_bill',data=df)
plt.show()
```

```python
sns.relplot(x='tips',y='total_bill',data=df,hue='sex')
plt.show()
```

```python
sns.relplot(x='tips',y='total_bill',data=df,hue='sex',style='smoker')
plt.show()
```

```python
sns.relplot(x='tips',y='total_bill',data=df,hue='sex',style='smoker',col='day')
plt.show()
```

```python
sns.relplot(x='tips',y='total_bill',data=df,hue='time',col='day')
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.relplot(x='tips',y='total_bill',data=df,col='day',kind='line',)
plt.show()
```

Relational == > scatter , line 

```python
sns.relplot(x='size',y='total_bill',data=df,col='day',kind='line',)
plt.show()
```

# Categorical plot

Strip chart

```python
sns.catplot(x='smoker',y='total_bill',data=df)
plt.show()
# Strip chart
```

```python
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex')
plt.show()
```

'strip', 'swarm', 'box', 'violin', 'boxen', 'point', 'bar', 'count' all form catpot chart

```python
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex',kind='swarm')
plt.title('swarm chart')
plt.show()
```

```python
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex',kind='violin')
plt.title('Violin chart')
plt.show()
```

```python
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex',kind='boxen')
plt.title('Boxen chart')
plt.show()
```

```python
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex',kind='bar')
plt.title('Bar chart')
plt.show()
```

```python
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex',kind='point')
plt.title('Point chart')
plt.show()
```

```python
sns.catplot(x='total_bill',data=df,hue='sex',kind='count')
plt.title('Count chart')
plt.show()
```
